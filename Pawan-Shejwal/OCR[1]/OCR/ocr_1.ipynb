```python
import tkinter as tk
from tkinter import filedialog
import cv2
import pytesseract
import pyttsx3

# Function to open an image file and perform OCR
def open_image():
    file_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.png *.jpg *.jpeg *.bmp *.gif *.tiff")])
    if file_path:
        img = cv2.imread(file_path)
        text = pytesseract.image_to_string(img)
        text_box.delete(1.0, tk.END)  # Clear any existing text
        text_box.insert(tk.END, text)  # Insert the extracted text into the text box
        speak_text(text)  # Speak the extracted text

# Function to convert and speak text
def speak_text(text):
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)
    engine.setProperty('volume', 1.0)
    engine.say(text)
    engine.runAndWait()

# Create the main application window
root = tk.Tk()
root.title("Image Text Extractor")

# Create a button to open an image
open_button = tk.Button(root, text="Open Image", command=open_image)
open_button.pack(pady=10)

# Create a text box to display the extracted text
text_box = tk.Text(root, height=10, width=40)
text_box.pack()

# Start the Tkinter main loop
root.mainloop()
```
