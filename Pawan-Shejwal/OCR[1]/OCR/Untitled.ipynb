```python
from tkinter import *
import pytesseract
from PIL import Image, ImageTk
import cv2
from tkinter import filedialog
import pyttsx3

win = Tk()
win.geometry("600x600+200+30")
win.resizable(False, False)
win.configure(bg='#1b407a')
w = 400
h = 300

color = "#581845"
frame_1 = Frame(win, width=600, height=320, bg=color)
frame_1.place(x=0, y=0)
frame_2 = Frame(win, width=600, height=320, bg=color)
frame_2.place(x=0, y=350)
frame_3 = Frame(win, width=600, height=80, bg=color)
frame_3.place(x=0, y=680)

v = Label(frame_1, width=w, height=h)
v.place(x=10, y=10)
cap = cv2.VideoCapture(0)

rgb_label = Label(frame_3, text="", bg=color)
rgb_label.place(x=10, y=10)

def speak_text(text_to_speak):
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)  # Adjust speech rate as needed
    engine.setProperty('volume', 1.0)  # Adjust volume as needed
    engine.say(text_to_speak)
    engine.runAndWait()

def save_image(image):
    file_path = filedialog.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png")])
    if file_path:
        image.save(file_path)
        print(f"Image saved as {file_path}")
    else:
        print("Image not saved.")

def take_copy(im):
    global rgb
    la = Label(frame_2, width=w-100, height=h-100)
    la.place(x=10, y=370)
    copy = im.copy()
    copy = cv2.resize(copy, (w-100, h-100))
    rgb_copy = cv2.cvtColor(copy, cv2.COLOR_BGR2RGB)
    image = Image.fromarray(rgb_copy)
    imgtk_copy = ImageTk.PhotoImage(image)
    la.configure(image=imgtk_copy)
    la.image = imgtk_copy  # Keep a reference to the PhotoImage

def select_img():
    global rgb, imgtk  # Keep references to rgb and imgtk
    _, img = cap.read()

    if img is not None and not img.size == 0:
        img = cv2.resize(img, (w, h))
        rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        image = Image.fromarray(rgb)
        imgtk = ImageTk.PhotoImage(image)
        v.configure(image=imgtk)
        v.image = imgtk
        show_rgb_values_on_image()
        v.after(10, select_img)
    else:
        print("Error: Webcam image is empty or invalid.")

def show_rgb_values(event):
    if 'img' in globals():
        x, y = event.x, event.y
        
        # Check if the x and y coordinates are within the valid range
        if 0 <= x < w and 0 <= y < h:
            pixel = img[y, x]
            rgb_label.config(text=f"RGB: {pixel[0]}, {pixel[1]}, {pixel[2]}")
        else:
            rgb_label.config(text="Out of bounds")

def process():
    global rgb
    image = Image.fromarray(rgb)
    save_image(image)
    
    pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
    img1 = cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR)  # Convert RGB image to BGR format
    cv2.imwrite('output.png', img1)
    Text = pytesseract.image_to_string('output.png')
    
    myfile = open('File.txt', 'w+')
    myfile.write(Text)
    myfile.close()
    print("text:", Text)
    
    OUTPUT = Toplevel(win)
    OUTPUT.title("EXTRACTED TEXT")
    OUTPUT.geometry('600x600')
    OUTPUT.config(bg="#fa8072")
    
    l6 = Label(OUTPUT, text="EXTRACTED TEXT", font=("Times New Roman bold", 27), bg='#fa8072', padx=20, pady=10).pack()
    l7 = Label(OUTPUT, text=Text, font=("Cambria", 13), bg='#fa8072', padx=20, pady=10).pack()
    l8 = Label(OUTPUT, text="The Above Text is Saved as an Editable (.txt) file which is saved in the current directory", bg="#fa8072", font=("Goudy old style", 12)).pack()
    
    speak_button = Button(OUTPUT, text="Speak Text", background="yellow", foreground="black", command=lambda: speak_text(Text))
    speak_button.place(x=250, y=250, width=100, height=50)

select_img()
snap = Button(win, text="capture", background="yellow", foreground='black', command=lambda: take_copy(rgb))
snap.place(x=450, y=150, width=60, height=50)
save = Button(win, text="save", background="yellow", foreground='black', command=lambda: save_image(Image.fromarray(rgb)))
save.place(x=450, y=400, width=60, height=50)
process1 = Button(win, text="process", background="yellow", foreground='black', command=process)
process1.place(x=450, y=550, width=60, height=50)

win.mainloop()
```

```python
from tkinter import *
from PIL import Image, ImageTk
import cv2
from tkinter import filedialog

print("project")
win = Tk()
win.geometry("600x600+200+30")
win.resizable(False, False)
win.configure(bg='#1b407a')
w, h = 400, 300
color = "#581845"

frame_1 = Frame(win, width=600, height=320, bg=color)
frame_1.place(x=0, y=0)

frame_2 = Frame(win, width=600, height=320, bg=color)
frame_2.place(x=0, y=350)

v = Label(frame_1, width=w, height=h)
v.place(x=10, y=10)
cap = cv2.VideoCapture(0)
rgb = None  # Initialize the 'rgb' variable

# Add a label to display the RGB value
rgb_label = Label(win, text="RGB Value: N/A", bg=color, fg="white", font=("Arial", 12))
rgb_label.pack(side=BOTTOM, pady=10)

def Save():
    global rgb
    if rgb is not None:
        image = Image.fromarray(rgb)
        file_path = filedialog.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png")])

        if file_path:
            image.save(file_path)
            print(f"Image saved as {file_path}")
        else:
            print("Image not saved.")
    else:
        print("No image to save.")

def take_copy(im):
    global rgb
    if rgb is not None:
        # Capture the current image displayed in the label
        copy = rgb.copy()
        image = Image.fromarray(copy)
        imgtk = ImageTk.PhotoImage(image)
        la = Label(frame_2, width=w-100, height=h-100)
        la.place(x=10, y=370)
        la.configure(image=imgtk)
        la.image = imgtk
        # Bind the mouse click event to get RGB value
        la.bind("<Button-1>", get_rgb_value)

def select_img():
    global rgb
    _, img = cap.read()

    if img is not None and not img.size == 0:
        img = cv2.resize(img, (w, h))
        rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        image = Image.fromarray(rgb)
        imgtk = ImageTk.PhotoImage(image)
        v.configure(image=imgtk)
        v.image = imgtk
    v.after(10, select_img)  # Keep capturing images

def get_rgb_value(event):
    global rgb
    if rgb is not None:
        x, y = event.x, event.y
        pixel_rgb = rgb[y, x]
        rgb_label.config(text=f"RGB Value: {pixel_rgb}")

select_img()
snap = Button(win, text="capture", background="yellow", foreground='black', command=lambda: take_copy(rgb))
snap.place(x=450, y=150, width=60, height=50)

save = Button(win, text="save", background="yellow", foreground='black', command=lambda: Save())
save.place(x=450, y=400, width=60, height=50)

win.mainloop()
cap.release()
```

```python
from tkinter import *
from PIL import Image, ImageTk
import cv2

win = Tk()
win.geometry("600x750+200+30")
win.resizable(False, False)
win.configure(bg='#1b407a')
w, h = 400, 300
color = "#581845"

frame_1 = Frame(win, width=600, height=320, bg=color)
frame_1.place(x=0, y=0)

frame_2 = Frame(win, width=600, height=320, bg=color)
frame_2.place(x=0, y=350)

frame_3 = Frame(win, width=600, height=80, bg=color)
frame_3.place(x=0, y=680)

# Create separate Label widgets for video feed and captured image
v_video = Label(frame_1, width=w, height=h)
v_video.place(x=10, y=10)

v_captured = Label(frame_2, width=w, height=h)
v_captured.place(x=10, y=370)

cap = cv2.VideoCapture(0)

rgb_label = Label(frame_3, text="", bg=color)
rgb_label.place(x=10, y=10)

def capture_image():
    global img
    _, img = cap.read()
    
    if img is not None and not img.size == 0:
        img = cv2.resize(img, (w, h))
        rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        image = Image.fromarray(rgb)
        imgtk = ImageTk.PhotoImage(image)
        v_captured.configure(image=imgtk)  # Display captured image in the v_captured Label
        v_captured.image = imgtk
        show_rgb_values_on_image()
    else:
        print("Error: Webcam image is empty or invalid.")

def save_image():
    if 'img' in globals():
        file_path = filedialog.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png")])
        if file_path:
            cv2.imwrite(file_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
            status_label.config(text="Image saved as " + file_path)

def process_image():
    if 'img' in globals():
        img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        cv2.imshow("Processed Image", img_gray)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

def show_rgb_values_on_image():
    if 'img' in globals():
        x, y = w // 2, h + 20  # Position to display RGB values on the image
        pixel = img[y, x]
        rgb_label.config(text=f"RGB: {pixel[0]}, {pixel[1]}, {pixel[2]}")

def show_rgb_values(event):
    if 'img' in globals():
        x, y = event.x, event.y
        
        # Check if the x and y coordinates are within the valid range
        if 0 <= x < w and 0 <= y < h:
            pixel = img[y, x]
            rgb_label.config(text=f"RGB: {pixel[0]}, {pixel[1]}, {pixel[2]}")
        else:
            rgb_label.config(text="Out of bounds")


def show_webcam():
    _, frame = cap.read()
    if frame is not None:
        frame = cv2.resize(frame, (w, h))
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image = Image.fromarray(rgb)
        imgtk = ImageTk.PhotoImage(image)
        v_video.configure(image=imgtk)  # Display video feed in the v_video Label
        v_video.image = imgtk
    win.after(10, show_webcam)  # Call show_webcam function after a delay (10ms)

snap = Button(win, text="Capture", background="yellow", foreground='black', command=capture_image)
snap.place(x=450, y=150, width=60, height=50)

save = Button(win, text="Save", background="yellow", foreground='black', command=save_image)
save.place(x=450, y=400, width=60, height=50)

process1 = Button(win, text="Process", background="yellow", foreground='black', command=process_image)
process1.place(x=450, y=550, width=60, height=50)

show_webcam()  # Start displaying the webcam video

win.mainloop()
cap.release()
```
