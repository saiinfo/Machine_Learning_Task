# day17:EvaluatingRegressionMOdelUsingRquaredAdjustRsquare

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#dataset =pd.read_csv("housedataset.csv")
dataset=pd.read_csv('housedataset.csv')
X=dataset.drop('price',axis=1)

y=dataset.drop('area',axis=1)

```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(dataset.area,dataset.price,color='red',marker='*')
```

```python
#spliting the dataset into training and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0) #state of random factors
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
plt.scatter(X,y,color="red",marker='*')
plt.plot(X,regressor.predict(X),color='green')
plt.title("Linear Regression")
plt.xlabel("Area")
plt.ylabel("price")
plt.show()
```

```python
#rquared score
from sklearn.metrics import r2_score
rsquared =regressor.score(X_test,y_test)
print(rsquared)
```

```python
r2_score(y,regressor.predict(X))
```

```python
#adjusted R Squared of the Model
n=len(dataset)
p=len(dataset.columns)-1
adjr=1-(1-rsquared)*(n-1)/(n-p-1)
print(adjr)
```
